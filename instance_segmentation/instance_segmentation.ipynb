# **Import the Required packages**

```python
import pixellib
from pixellib.instance import instance_segmentation
import cv2
import numpy as np
from PIL import Image
from IPython.display import Image as img
from matplotlib import pyplot as plt
```

# **load the model**

```python
segment_image = instance_segmentation()
segment_image.load_model("/drive/My Drive/pixillib/segmentation/instance_segmentation/mask_rcnn_coco.h5") 
```

# **start the instance segmentation**

```python
sample_img = "/drive/My Drive/sample.jpg"
segment_image.segmentImage(sample_img, output_image_name = "output_img.jpg")
plt.imshow(Image.open("output_img.jpg"))
```

```python
segment_image.segmentImage(sample_img, output_image_name = "output_img.jpg", show_bboxes = True)
plt.imshow(Image.open("output_img.jpg"))
```

```python
segmask, output = segment_image.segmentImage(sample_img)
cv2.imwrite("output_img.jpg", output)
print(output.shape)
plt.imshow(Image.open("output_img.jpg"))
```

```python
segmask, output = segment_image.segmentImage(sample_img, show_bboxes= True)
cv2.imwrite("output_img.jpg", output)
print(output.shape)
plt.imshow(Image.open("output_img.jpg"))
```
