# **Import the Required packages**

```python
import pixellib
from pixellib.semantic import semantic_segmentation

from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
from IPython.display import Image as img
import cv2
```

Download the pretrained model [here](https://github.com/ayoolaolafenwa/PixelLib/releases/download/1.3/deeplabv3_xception65_ade20k.h5)

# **load the model**

```python
segment_image = semantic_segmentation()
segment_image.load_ade20k_model("/drive/My Drive/pixillib/segmentation/sematic_150/deeplabv3_xception65_ade20k.h5")
```

# **start the sematic segmentation**

```python

sample_img = "/drive/My Drive/sample.jpg"
segment_image.segmentAsAde20k(sample_img, output_image_name= "output_img.jpg")
plt.imshow(Image.open("output_img.jpg"))
```

```python
segment_image.segmentAsAde20k(sample_img, output_image_name = "output_img.jpg", overlay = True)
plt.imshow(Image.open("output_img.jpg"))
```

```python
segmap, output = segment_image.segmentAsAde20k(sample_img)
cv2.imwrite("output_img.jpg", output)
print(output.shape)
plt.imshow(Image.open("output_img.jpg"))
```

```python
segmap, seg_overlay = segment_image.segmentAsAde20k(sample_img, overlay = True)
cv2.imwrite("output_img.jpg", seg_overlay)
print(seg_overlay.shape)
plt.imshow(Image.open("output_img.jpg"))
```
