# **Import the Required packages**

```python
import pixellib
from pixellib.semantic import semantic_segmentation
import cv2
import numpy as np
from PIL import Image
from IPython.display import Image as img
from matplotlib import pyplot as plt
```

Download the pretrained model [here](https://github.com/ayoolaolafenwa/PixelLib/releases/download/1.1/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5)

# **load the model**

```python
sample_img = "/drive/My Drive/sample.jpg"
segment_image = semantic_segmentation()
segment_image.load_pascalvoc_model("/drive/My Drive/pixillib/segmentation/sematic_segmentation/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5") 
```

# **start the sematic segmentation**

```python
segment_image.segmentAsPascalvoc(sample_img, output_image_name = "output_img.jpg")
plt.imshow(Image.open("output_img.jpg"))
```

```python
segment_image.segmentAsPascalvoc(sample_img, output_image_name = "output_img.jpg", overlay = True)
plt.imshow(Image.open("output_img.jpg"))
```

```python
output, segmap = segment_image.segmentAsPascalvoc(sample_img)
cv2.imwrite("output_img.jpg", output)
print(output.shape)
plt.imshow(Image.open("output_img.jpg"))
```

```python
segmap, segoverlay = segment_image.segmentAsPascalvoc(sample_img, overlay= True)
cv2.imwrite("output_img.jpg", segoverlay)
print(segoverlay.shape)
plt.imshow(Image.open("output_img.jpg"))
```
